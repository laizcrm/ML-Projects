**K Means Clustering**


É uma técnica de aprendizado de máquina não supervisionada usada para identificar clusters de objetos de dados (agrupamento dos dados a partir de características similares) em um conjunto de dados, considerando um número *K* mais efetivo (n possibilidades diferentes- exs.: número de vizinho, PCA, soma dos erros quadrados) de clusters. Para cada cluster, o algoritmo calcula a média dos vetores de cada ponto do cluster.

*   Clusters:

Conjunto de técnicas usadas para particionar dados. Clusters são  definidos como grupos de objetos de dados que são mais semelhantes a outros objetos. 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#plot sem usar o plt.show():
%matplotlib inline      
!pip install scikit-learn
```

```python
#make blobs função que gera dados artificiais aleatórios baseados em clusters 
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4, cluster_std=1.0,random_state=50)
```

```python
print(type(data))
#array de 200 linhas com 2 col (param)
print(data[1]) #classe que o dado pertence com 4 centróides
```

```python
#Análise de dados

plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
#c mostra a cor conforme cada classe 


```

```python
#Construção do Modelo

#Biblioteca responsável por separar os dados em teste e treino e associar os parâmetros 
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)

#Como é um modelo NÃO SUPERVISIONADO, não é necessária a quebra em treino e teste. 
#Este modelo não COMPARA e sim CLASSIFICA (não tem a avaliação de performance de modelo) 
```

```python
kmeans.fit(data[0])
```

```python
#centróides dos clusters 
kmeans.cluster_centers_

#valores para centróides
kmeans.labels_
```

```python
#figuras em 2 eixos, com 1 linha, 2 col e compartilhem os valores de y 

f,(ax1,ax2)= plt.subplots(1,2,sharey=True)

#eixo1 :esquerda
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

#eixo2 :direita
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')

```
